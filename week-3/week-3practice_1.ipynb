```python
def printinfo( arg1, *vartuple ):
    #test it again
    print ("Output is: ")
    print (arg1)
    for var in vartuple:
        print (var)
        return;       #If u change the indention of return it wiil change the results which in this case it will include 50
printinfo( 10 );
printinfo( 70,60,50 );
```
